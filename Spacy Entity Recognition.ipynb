```python
f = feedparser.parse(url)
# feedparser.parse extracts the required objects from the url
```

```python
import pandas as pd
```

```python
data = {}
# Creating an empty dictionary
```

```python

for entry in f.entries:
    
    data.setdefault('title',[])
    data.setdefault('summary',[])
    data.setdefault('published',[])
    data.setdefault('link',[])
    data['title'].append(entry.title)
    data['summary'].append(entry.summary)
    data['published'].append(entry.published)
    data['link'].append(entry.link)
# Appending the required data to the empty dictionary
```

```python
df = pd.DataFrame(data)
# Pushing the data to a dataframe
```

```python
a = df.link[3]
```

```python
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
g_feed = url_to_string(a)
article = nlp(g_feed)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```
